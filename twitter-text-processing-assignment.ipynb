```python
#stop words
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    new_text = []
    for words in text.split():
        if words in stopwords.words('english'):
            new_text.append("")
        else:
            new_text.append(words)
    buffer=" ".join(new_text)
    return " ".join(buffer.split())

df_text['text']=df_text['text'].apply(remove_stopwords)
df_text['text'].head()

```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
df_text['text']=df_text['text'].apply(remove_stopwords)
df_text['text'].head()
```

```python
chatwords_dict = {}
chatwords_list = []
for line in chat_words_str.split('\n'):
    if line!='':
       chat_key=line.split('=')[0]
       chat_expanded=line.split('=')[1]
       chatwords_list.append(chat_key)
       chatwords_dict[chat_key] = chat_expanded
chatwords_list=set(chatwords_list)
print(chatwords_dict)
def remove_chatwords(text):
    new_text = []
    for w in text.split():
        if w.upper() in chatwords_list:
            new_text.append(chatwords_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

text="WTG"
print(remove_chatwords(text))
df_text['text']=df_text['text'].apply(remove_chatwords)
df_text['text'].head()
```

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

df_text['text']=df_text['text'].apply(remove_urls)
df_text['text'].head()
```

```python
def remove_emoji(text):
    clean_text=emoji.demojize(text)
    return clean_text
remove_emoji(text)
df_text['text']=df_text['text'].apply(remove_emoji)

import re
def remove_emojis_manually(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    clean_text = emoji_pattern.sub(r'', text)
    return clean_text

df_text['text']=df_text['text'].apply(remove_emojis_manually)
```

```python
from nltk.stem import WordNetLemmatizer

def lammatization(text):
    words=text.split()

    lemmetizer=WordNetLemmatizer()

    lemetized_word=" ".join([lemmetizer.lemmatize(word) for word in words])
    return " ".join(lemetized_word.split())
```

```python
df_text['lammatization'] = df_text['text'].apply(lammatization)
df_text['lammatization'].head()
```

```python
OneHotEncoder
# Advantage
# It's relatively straightforward to implement and enables us to apply machine-learning algorithms to data with categorical columns
# One-hot encoding has the advantage that the result is binary rather than ordinal and that everything sits in an orthogonal vector space
# Disadvantage
# The problem is that it increases dimensionality so training becomes slower and more complex
# When we one-hot encode categorical data, we unknowingly introduce perfect multicollinearity
# Increase in dimensionality, after adding several columns based on categorical variables which may result in an increase in the computational cost
```

```python
# Bag of words
# Advantage
# One advantage is that it provides a feature representation of free-form text that can be used by machine learning algorithms for processing language
# Captures the essence of the document by representing it as a set of its constituent words.	
# Robust to noise, as representation of the document is based only on the occurrence of its constituent words.	
# Disadvantage
# It does not take into account the semantics of the language, as it is based on a system of binary logic
# The bag-of-words approach may not capture the syntactic and layout information present in the text
# Unsuitable for tasks, where the order of words is important.
```

```python
# N-Gram
# Disadvantage
# some n-grams may not occur frequently or at all in the training data, resulting in low or zero probabilities

# Advantage
# capture information about the structure of words and phrases in a text data set.
# N-grams can provide features that are insensitive to the ordering of words in a sentence.
# N-grams can be used to create features that are more robust to spelling errors and other types of noise in the data
# Finally, n-grams can help to reduce the dimensionality of the feature space
```

```python
# TD-IDF
# Advantages:
# - Easy to compute
# - You have some basic metric to extract the most descriptive terms in a document
# - You can easily compute the similarity between 2 documents using it

# Disadvantages:
# - TF-IDF is based on the bag-of-words (BoW) model, therefore it does not capture position in text, semantics, co-occurrences in different documents, etc.
# - For this reason, TF-IDF is only useful as a lexical level feature
# - Cannot capture semantics (e.g. as compared to topic models, word embeddings)
```
